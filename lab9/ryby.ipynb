```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Reverse colors

```python
galia = cv2.imread("./school.png")
galia = 255 - galia

h, w, _ = galia.shape

print(h)

e_letter = cv2.imread("./fish1.png")
e_letter = 255 - e_letter

galia = cv2.cvtColor(galia, cv2.COLOR_BGR2GRAY)
e_letter = cv2.cvtColor(e_letter, cv2.COLOR_BGR2GRAY)

plt.imshow(galia, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()

plt.imshow(e_letter, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

```python
res = cv2.matchTemplate(galia, e_letter, cv2.TM_CCOEFF_NORMED)
```

```python
tsh = 0.65
ys, xs = np.where(res > tsh)

h_t, w_t = e_letter.shape
print(h_t, w_t)

galia_marked = galia.copy()

for x, y in zip(xs, ys):
    cv2.rectangle(galia_marked, (x, y), (x + w_t, y + h_t), 0, 1)

plt.imshow(galia_marked, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

```python
import numpy as np
import time
```

```python
def find_pivot(matrix, col, start_row):
    our_col = abs(matrix[start_row:, col])
    return our_col.argmax() + start_row
        
```

```python
def Gauss_Jordan(matrix, b):
    
    matrix_save = matrix.copy()
    b_save= b.copy()
    
    t_s = time.time()
    
    n = matrix.shape[0]
    
    for i in range(n):
        max_idx = find_pivot(matrix, i, i)
        matrix[[i, max_idx]] = matrix[[max_idx, i]]
        b[[i, max_idx]] = b[[max_idx, i]]
            
        pivot = matrix[i, i]
        
        if abs(pivot) < 1e-10:
            continue
    
        for j in range(n):
            if j != i:
                factor = matrix[j, i] / pivot
                matrix[j, i:] -= factor * matrix[i, i:]
                b[j] -= factor * b[i]
    
    x_counted = b / np.diag(matrix)
    
    t_e = time.time()
    t_s_l = time.time()
    
    x_numpy = np.linalg.solve(matrix_save, b_save)
    test_numpy = np.allclose(x_counted, x_numpy)
    
    t_e_l = time.time()
    
    return test_numpy, t_e-t_s, t_e_l-t_s_l
                       
```

```python
for i in range(10):
    dimension = np.random.randint(501, 1500)
    
    matrix = np.random.rand(dimension, dimension)
    b = np.random.rand(dimension)
    
    val, time_i, time_lib = Gauss_Jordan(matrix, b)
    
    print(f"Test {i+1}, dimesion {dimension},  output {val}, time implemented {time_i:.4f}, time library {time_lib:.4f}, diff {time_i-time_lib:.4f}"
    )
```

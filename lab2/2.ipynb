```python
import numpy as np
```

```python
def LU(matrix):
    
    A = matrix.copy()
    
    n = matrix.shape[0]
    
    for i in range(n):            
        for j in range(i+1, n):
            matrix[j, i] = matrix[j, i] / matrix[i, i]
            matrix[j, i+1:] -= matrix[j, i] * matrix[i, i+1:]
            
            
    #Wyliczenie |A-LU| i sprawdzenie czy się zerują wartości
    matrix_val = np.zeros((n, n))
    matrix_counted = A - (np.tril(matrix, k=-1) + np.eye(n)) @ np.triu(matrix)
                
    return np.allclose(matrix_val, matrix_counted)
```

```python
for i in range(10):
    dimension = np.random.randint(501, 1500)
    
    matrix = np.random.rand(dimension, dimension)
    
    print(f"Test {i+1}, dimesion {dimension}, output {LU(matrix)}"
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

Load my image

```python
image = Image.open("lena_gray.bmp")
image_array = np.array(image)
```

```python
U, sigma, V = np.linalg.svd(image_array)

sum = np.zeros_like(image_array, dtype=float)

for k in range(10):
    uk = U[:, k]
    vk = V[k, :]

    uk = uk.reshape(-1, 1)
    vk = vk.reshape(1, -1)

    sum += sigma[k] * (uk @ vk)

img = Image.fromarray(sum)
img.show()
```

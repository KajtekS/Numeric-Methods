Metoda potęgowa

```python
def power_method(A):
    n, m = A.shape
    x = np.ones((n, 1))
    lambda_o = 0

    for i in range(100):
        x_p = A @ x
        lambda_n = np.linalg.norm(x_p, np.inf)

        x_p /= lambda_n

        if abs(lambda_n - lambda_o) < 1e-5:
            return lambda_n, x_p, i

        lambda_o = lambda_n
        x = x_p
            
        x = x_p
    
    
    
    return lambda_o, x, 100 
```

```python
A = np.random.rand(100, 100)
A = (A + A.T) / 2
```

```python
lambda_o, x, i = power_method(A)
```

```python
print(lambda_o)
print(x)
```

```python
w, v = np.linalg.eig(A)
max_index = np.argmax(w)
max_val = w[max_index]
max_vec = v[:, max_index]

print(abs(max_val) - lambda_o)

max_vec_normalized = max_vec / np.linalg.norm(max_vec, np.inf)

print(
    np.linalg.norm(np.abs(max_vec_normalized) - np.abs(x.flatten()), np.inf)
)
```

```python
def ray(A):
    n, m = A.shape
    x = np.ones((n, 1))
    lambda_o = 0

    sigma = (x.T @ A @ x) / (x.T @ x)

    B = A - sigma * np.eye(n)

    lambda_o = 0
    sigma = 0

    for j in range(100):
        B = A - sigma * np.eye(n)

        x_p = np.linalg.solve(B, x)
        lambda_n = np.linalg.norm(x_p, np.inf)

        x_p /= lambda_n

        if abs(lambda_n - lambda_o) < 1e-5:
            lambda_o = lambda_n
            x = x_p
            break

        lambda_o = lambda_n
        x = x_p

        sigma = (x.T @ A @ x) / (x.T @ x)
        B = A - sigma * np.eye(n)

    return sigma, x
```
